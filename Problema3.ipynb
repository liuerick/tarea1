```python
# Problema 3: Detector de puntaje gamer

puntaje1 = int(input("Ingresa primer puntaje: "))
puntaje2 = int(input("Ingresa segundo puntaje: "))

suma = puntaje1 + puntaje2

if suma < 50:
    print("Resultado: Nivel principiante")
elif suma < 80:
    print("Resultado: Nivel intermedio")
else:
    print("Resultado: Nivel pro")
```
